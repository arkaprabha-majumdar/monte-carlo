```
import numpy as np
import pandas as pd
```

```
def casino_roll():
    roll = np.random.randint(1,100)

    #if roll <= 50 or roll == 100:
        #print('rolled %d. you lose. House wins! Play again?!'%roll)
    if 50 < roll < 100:
        #print('you win! On a roll !')
        return True
    else:
        return False
```

```
casino_roll()
```

```
def deal_bet(funds,initial_bet,bet_count):
    bet = 0
    funds_total = 0
    while bet < bet_count:
        if casino_roll():
            funds += initial_bet
        else:
            funds -= initial_bet

        bet += 1
        #print('Funds : %d'%funds)
        funds_total += funds
    print('average win/loss : %f'%(10000-funds_total/bet_count))
    return (10000-funds_total/bet_count)
```

```
x=0
avg_list=[]
while x<100:
  avg_list.append(deal_bet(10000,100,100))
  x+=1
```

```
import matplotlib.pyplot as plt

avg_list = np.array(avg_list).astype(int)
plt.figure(figsize=(20,20))
plt.plot(avg_list)
plt.axhline(np.array(avg_list).mean(), color='k', linestyle='dashed', linewidth=1)

max_val = np.amax(np.array(avg_list).astype(int))
max_index = np.where(avg_list == np.amax(np.array(avg_list).astype(int)))
min_val = np.amin(np.array(avg_list).astype(int))
min_index = np.where(avg_list == np.amin(np.array(avg_list).astype(int)))
```
